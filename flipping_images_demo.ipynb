```python
from PIL import Image, ImageOps

# Load your image
image_path = 'image-to-flip.jpg'
image = Image.open(image_path)

# Flip the image horizontally
h_flipped_image = ImageOps.mirror(image)

# Flip the image vertically
v_flipped_image = ImageOps.flip(image)

# Display the original and flipped images
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(h_flipped_image)
axs[1].set_title('Horizontally Flipped')
axs[1].axis('off')

axs[2].imshow(v_flipped_image)
axs[2].set_title('Vertically Flipped')
axs[2].axis('off')

plt.show()
```
